```python
# Importing the required libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Downloading contents of the web page
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
data = requests.get(url).text
```

```python
# Creating BeautifulSoup object
soup = BeautifulSoup(data, 'html.parser')
```

```python
# Verifying tables and their classes
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
# Creating list with all tables
tables = soup.find_all('table')

#  Looking for the table with the classes 'wikitable' and 'sortable'
table = soup.find('table', class_='wikitable sortable')
```

```python
table
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['Country or Region', 'Population', '% of the world', 'Date', 'Source'])

# Collecting Ddata
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    
    if(columns != []):

        cor = columns[0].text.strip()
        pop = columns[1].text.strip()
        percent = columns[2].text.strip()
        date = columns[3].text.strip()
        source = columns[4].text.strip()

        df = df.append({'Country or Region': cor,  'Population': pop, '% of the world': percent, 'Date': date, 'Source': source}, ignore_index=True)
```

```python
df.drop(index=0, axis=0, inplace=True)
```
